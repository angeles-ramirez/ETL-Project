```python
# United Nations: Extraction
UN_data_file = "../Selected Data/UN_Data_Country_Profiles/country_profile_variables.csv"
UN_data_df = pd.read_csv(UN_data_file, encoding="ISO-8859-1")
UN_data_df.head()
```

```python

# Set Index
UN_data_df.set_index('country', inplace = True)
# Display all columns
UN_data_df.columns
```

```python
# United Nations: Transformation
UN_data_cols = ["Region", 
                "Population in thousands (2017)", 
                "Population density (per km2, 2017)", 
                "Sex ratio (m per 100 f, 2017)", 
                "GDP: Gross domestic product (million current US$)", 
                "GDP per capita (current US$)", 
                "Unemployment (% of labour force)", 
                "Population growth rate (average annual %)", 
                "Fertility rate, total (live births per woman)"]
UN_data_transformed= UN_data_df[UN_data_cols].copy()

UN_data_transformed = UN_data_transformed.rename(columns={
                                                        "Population in thousands (2017)": "pop_in_1000s", 
                                                        "Population density (per km2, 2017)": "pop_dens_per_sqkm", 
                                                        "Sex ratio (m per 100 f, 2017)": "males_per_100_fem", 
                                                        "GDP: Gross domestic product (million current US$)": "GDP_in_mil_dol", 
                                                        "GDP per capita (current US$)": "GDP_per_cap_in_mil", 
                                                        "Unemployment (% of labour force)": "unemploy_perct",
                                                        "Population growth rate (average annual %)": "pop_growth_rate_perct",
                                                        "Fertility rate, total (live births per woman)": "live_births_per_woman"
                                                        })
UN_data_transformed.head()
```

```python
#Economic Freedom: Extraction
raw_file = "../Selected Data/economic_freedom_index2019_data.csv"
free_eco_df = pd.read_csv(raw_file, encoding="ISO-8859-1")
free_eco_df.head()
```

```python
#Economic Freedom: Transformation
free_eco_df.set_index('CountryID', inplace = True)
free_eco_df.head()
```

```python
#Economic Freedom: Transformation

free_eco_df.columns=free_eco_df.columns.str.replace('(','')
free_eco_df.columns=free_eco_df.columns.str.replace(')','')
free_eco_df.columns=free_eco_df.columns.str.replace('%','')
free_eco_df.columns=free_eco_df.columns.str.replace(',','')
free_eco_df.columns=free_eco_df.columns.str.replace(r"[\"\',]", '')
free_eco_df.columns =free_eco_df.columns.str.rstrip()

free_eco_df = free_eco_df[free_eco_df.columns.drop('Country')]

free_eco_df.columns=free_eco_df.columns.str.replace(' ','_')
free_eco_df.columns

```

```python
#Economic Freedom: Transformation

free_eco_df= free_eco_df.rename(columns={  
            'Tax Burden of GDP' :  'Tax Burdenof GDP', 
            'Public Debt  of GDP': 'Public Debtof GDP', 
            'Govt Expenditure  of GDP' :'Govt Expenditure of GDP'
            })

free_eco_df.columns
```

```python
print(free_eco_df.dtypes)
```

```python
#Economic Freedom: Transformation


free_eco_df2 = free_eco_df.replace('N/A',0) 
free_eco_df3 = free_eco_df2.fillna(0)
# CLEAN SEVERAL COLUMNS 
# make a list of columns that need cleaning = 'cols'
cols = ['Population_Millions', 'Unemployment', 'GDP_Billions_PPP', 'GDP_per_Capita_PPP' ,'FDI_Inflow_Millions']
# remove  $ dollars sign and commas # from 3 columns  
# this dataframe is called  ' free_eco_df3 ' 
# this is how you remove parethesis   '\( \)':  YOU MUST HAVE A SPACE BTWN BACKSLASHES !!! OR WONT WORK you cant write  ')' : '' 
free_eco_df3[cols] = free_eco_df3[cols].replace({'2009':'',  '2014':'',  '2015': '',   '2016':'',    'CHF' :'',  'ppl.' :'', 'est.' :'', '\$': '',   ',': '' }, regex=True) 
 
free_eco_df3.dtypes
```

```python
free_eco_df3[cols] = free_eco_df3[cols].replace( {  '\( \)': '' }, regex=True ) 
free_eco_df3[cols] = free_eco_df3[cols].replace( {  '\(\)': '' }, regex=True ) 
```

```python
free_eco_df3[cols] = free_eco_df3[cols].apply(pd.to_numeric, errors='coerce')
#run dtypes to verify converion 
free_eco_df3.dtypes
```

```python
free_eco_df3.to_csv("free_eco_clean.csv", encoding="utf-8", index=True, header=True)
free_eco_df3
```
